## This F-test for features selection

```python
from sklearn.feature_selection import SelectKBest, f_classif

def my_feature_selection(X_train, y_train, X_test, no_of_features="all"):
    skb = SelectKBest(score_func=f_classif, k =no_of_features)
    skb.fit(X_train, y_train)
    X_train_fs = skb.transform(X_train)
    X_test_fs = skb.transform(X_test)
    return X_train_fs, X_test_fs, skb
```

```python
print("Shape of X_train:", X_train.shape)

# Call my_feature_selection and inspect X_train before fitting SelectKBest
X_train_fs, X_test_fs, skb = my_feature_selection(X_train, y_train, X_test, 20)
print("X_train_fs shape:", X_train_fs.shape)  
```
